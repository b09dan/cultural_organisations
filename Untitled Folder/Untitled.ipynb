```python
category = [] 
name = []
shortDescription=[]
description=[]
organization=[]
local=[]
start=[]
end=[]
cat_org=[]
id_org=[]
id_event=[]
```

```python
%%time

emoji_pattern = re.compile("["
                            u"\U0001F600-\U0001F64F"  # emoticons
                            u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                            u"\U0001F680-\U0001F6FF"  # transport & map symbols
                            u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            u"\xa0"
                           "\n"
                           "]+", flags=re.UNICODE)

for i in np.arange(0,total_events+100-total_events%100,100)[759:]:
    
    url = "https://all.culture.ru/api/2.2/events?limit=100&offset=" +str(i)+"&fields=_id,altName,name,ageRestriction,category,seances,shortDescription,places,description,organization,.json"
    response = urllib.request.urlopen(url)
    content = response.read()
    data = json.loads(content.decode("utf8"))
    print(i, data['events'][0]['name'])
    max_seanses=[]
    min_seanses=[]
    
    for event in data['events']:
        name.append(event['name']) 
        category.append(event['category']['name']) 
        shortDescription.append(emoji_pattern.sub(r'', BeautifulSoup(event['shortDescription'], "html.parser").get_text())) 
        description.append(emoji_pattern.sub(r'', BeautifulSoup(event['description'], "html.parser").get_text())) 
        for seans in event['seances']:
            max_seanses.append(seans['end'])
            min_seanses.append(seans['start'])
        organization.append(event['organization']['locale']['name'])
        try:
            category_org=event['places'][0]['category']['name']
        except:
            category_org="Unknown"
        cat_org.append(category_org)
        end.append(time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(np.max(max_seanses)/1000.0)))
        start.append(time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(np.min(min_seanses)/1000.0)))
        id_org.append(event['organization']['_id'])
        id_event.append(event['_id'])
        
        try:
            with conn.cursor() as cursor:
                sql = '''
                INSERT INTO `events` 
                (`category`, 
                `description`,
                `organization_category`, 
                `end`, 
                `start`,
                `id_org`,
                `id_event`, 
                `name`,
                `organization_local`, 
                `shortDescription`
                )
                VALUES (%s, %s, %s,%s, %s, %s, %s, %s, %s, %s);
                '''
                cursor.execute(sql, 
                               (event['category']['name'], 
                                emoji_pattern.sub(r'', BeautifulSoup(event['description'], "html.parser").get_text()),
                                category_org,
                                str(time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(np.max(max_seanses)/1000.0))),
                                str(time.strftime('%Y-%m-%d %H:%M:%S', time.gmtime(np.min(min_seanses)/1000.0))),
                                event['organization']['_id'],
                                event['_id'],                                            
                                event['name'],                                            
                                event['organization']['locale']['name'],
                                emoji_pattern.sub(r'', BeautifulSoup(event['shortDescription'], "html.parser").get_text())
                               ))
                conn.commit()
        finally:
            print('finally')
```
